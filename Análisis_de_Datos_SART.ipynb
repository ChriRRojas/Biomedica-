```python
import pandas as pd
import os


path = '/content/drive/MyDrive/Participantes_SART'


archivos = [f for f in os.listdir(path) if f.endswith('.csv')]


lista_dfs = []


for archivo in archivos:
    df = pd.read_csv(os.path.join(path, archivo), encoding='ISO-8859-1')
    lista_dfs.append(df)


df_unificado = pd.concat(lista_dfs, ignore_index=True)


print(df_unificado.head())
```

```python

df_unificado.to_csv('/content/df_unificado.csv', index=False)
```

```python
!pip install matplotlib seaborn
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df_unificado = pd.read_csv('/content/drive/MyDrive/Prueba SART proto/participantes unificados.csv')

```

```python
# Crear el boxplot
plt.figure(figsize=(8, 4))  # Tamaño de la figura
sns.boxplot(data=df_unificado, x='RT')  # Cambia 'Edad' por la columna que quieras graficar

# Agregar título y etiquetas
plt.title('Tiempo de respusta ')
plt.xlabel('nose jsjsjs')

# Mostrar la gráfica
plt.show()
```

```python
# Crear un histograma para una columna (por ejemplo, 'edad')
plt.figure(figsize=(10, 6))  # Tamaño de la figura
plt.hist(df_unificado['RT'], bins=20, edgecolor='black')  # Cambia 'edad' por la columna que desees graficar

# Agregar título y etiquetas
plt.title('Histograma de RT (tiempo de respuesta)')
plt.xlabel('RT')
plt.ylabel('Frecuencia')

# Mostrar la gráfica
plt.show()
```

```python
#Crear un KDE Plot para una columna (por ejemplo, 'edad')
plt.figure(figsize=(10, 6))  # Tamaño de la figura
sns.kdeplot(df_unificado['RT'], shade=True)  # Cambia 'edad' por la columna que desees graficar

# Agregar título y etiquetas
plt.title('KDE Plot de Edad en los Participantes')
plt.xlabel('Edad')
plt.ylabel('Densidad')

# Mostrar la gráfica
plt.show()
```

```python
# Crear un violin plot para una columna (por ejemplo, 'edad')
plt.figure(figsize=(10, 6))  # Tamaño de la figura
sns.violinplot(data=df_unificado, x='RT')  # Cambia 'edad' por la columna que desees graficar

# Agregar título y etiquetas
plt.title('Violin Plot de Edad en los Participantes')
plt.xlabel('Edad')

# Mostrar la gráfica
plt.show()
```

```python
# Crear un gráfico de dispersión (scatter plot) entre dos columnas (por ejemplo, 'edad' y 'presion')
plt.figure(figsize=(10, 6))  # Tamaño de la figura
sns.scatterplot(data=df_unificado, x='Edad', y='RT')  # Cambia 'edad' y 'presion' por las columnas que desees

# Agregar título y etiquetas
plt.title('Gráfico de Dispersión de Edad vs RT')
plt.xlabel('Edad')
plt.ylabel('RT')

# Mostrar la gráfica
plt.show()
```
